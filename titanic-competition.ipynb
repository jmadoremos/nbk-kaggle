```python
import tensorflow as tf
import pandas as pd
from matplotlib import pyplot
```

```python
train_data_all = pd.read_csv(".dataset/titanic-train.csv")
train_data_all.head()
train_data_all.info()
```

```python
test_data_all = pd.read_csv(".dataset/titanic-test.csv")
test_data_all.head()
```

```python
train_data_one_hot = pd.get_dummies(train_data_all)
train_data_one_hot.head()
```

```python
train_data_one_hot.describe()
```

```python
input_columns = [
    "Pclass",
    "SibSp",
    "Parch"
]
label_columns = [
    "Survived"
]

input_count = len(input_columns)
label_count = len(label_columns)

test_data_sample = train_data_one_hot.sample(frac=0.2, random_state=162)
test_data_input = test_data_sample.filter(input_columns)
test_data_label = test_data_sample.filter(label_columns)

train_data_sample = train_data_one_hot.drop(test_data_sample.index)
train_data_input = train_data_sample.filter(input_columns)
train_data_label = train_data_sample.filter(label_columns)

print("i: {}, l: {}".format(input_count, label_count))
```

```python
statistics = tf.placeholder(tf.float32, [None, input_count])
classification = tf.placeholder(tf.float32, [None, label_count])

W_input = tf.Variable(tf.zeros([input_count, label_count]))
b_input = tf.Variable(tf.zeros([label_count]))
prediction_input = tf.nn.softmax(tf.matmul(statistics, W_input) + b_input)

error_rate = tf.reduce_mean(-tf.reduce_sum(classification * tf.log(prediction_input), reduction_indices = [1]))
optimizer = tf.train.GradientDescentOptimizer(0.001).minimize(error_rate)
```

```python
iterations = 10000

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

correct_prediction = tf.equal(tf.argmax(prediction_input, 1), tf.argmax(classification, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

for step in range(iterations):
    sess.run(optimizer, feed_dict={
        statistics: train_data_input,
        classification: train_data_label
    })
    
    if step % (iterations * 0.1) == 0:
        print("accuracy: {}".format(
            sess.run(accuracy, {
                statistics: test_data_input,
                classification: test_data_label
            })
        ))

print("-------")
check_sample = [[3, 1, 0]]
sess.run(prediction_input, { statistics: check_sample })
```
